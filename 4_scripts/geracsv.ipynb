```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
navegador = webdriver.Chrome()
```

```python
navegador.get('https://finance.yahoo.com/lookup')
```

```python
navegador.maximize_window()
```

```python
nome = '#\\30  > td:nth-child(2) > span > div'
```

```python
navegador.find_element(By.CSS_SELECTOR, nome).text
```

```python
preco = '#\\30  > td:nth-child(3) > span'
navegador.find_element(By.CSS_SELECTOR, preco).text
```

```python
change = '#\\30  > td:nth-child(4) > span > fin-streamer'
navegador.find_element(By.CSS_SELECTOR, change).text
```

```python
lista_change = []
for change in range(10):
    try:
        # Seletor CSS para números de 0 a 9
        seletor = f'#\\3{change}  > td:nth-child(4) > span > fin-streamer'
        dado_change = navegador.find_element(By.CSS_SELECTOR, seletor).text
        lista_change.append(dado_change)
        print(dado_change)
    except:
        pass
        

for change in range(10, 100):
    try:
        # Seletor CSS para números de 10 em diante
        seletor = f'#\\3{empresas // 10} {empresas % 10}  > td:nth-child(4) > span > fin-streamer'
        dado_change = navegador.find_element(By.CSS_SELECTOR, seletor).text
        lista_change.append(dado_change)
        print(dado_change)
    except:
        pass
```

```python
change2 = '#\\31  > td:nth-child(5) > span > fin-streamer > span'
navegador.find_element(By.CSS_SELECTOR, change2).text
```

```python
lista_change2 = []
for change2 in range(10):
    try:
        # Seletor CSS para números de 0 a 9
        seletor = f'#\\3{change2} > td:nth-child(5) > span > fin-streamer > span'
        dado_change2 = navegador.find_element(By.CSS_SELECTOR, seletor).text
        lista_change.append(dado_change2)
        print(dado_change2)
    except:
        pass
        

for change2 in range(10, 100):
    try:
        # Seletor CSS para números de 10 em diante
        seletor = f'#\\3{change2 // 10} {change2 % 10}  > td:nth-child(5) > span > fin-streamer > span'
        dado_change2 = navegador.find_element(By.CSS_SELECTOR, seletor).text
        lista_change2.append(dado_change2)
        print(dado_change2)
    except:
        pass
```

```python
lista_empresas = []
for empresas in range(10):
    try:
        # Seletor CSS para números de 0 a 9
        seletor = f'#\\3{empresas} > td:nth-child(2) > span > div'
        dado_empresa = navegador.find_element(By.CSS_SELECTOR, seletor).text
        lista_empresas.append(dado_empresa)
        print(dado_empresa)
    except:
        pass
        

for empresas in range(10, 100):
    try:
        # Seletor CSS para números de 10 em diante
        seletor = f'#\\3{empresas // 10} {empresas % 10} > td:nth-child(2) > span > div'
        dado_empresa = navegador.find_element(By.CSS_SELECTOR, seletor).text
        lista_empresas.append(dado_empresa)
        print(dado_empresa)
    except:
        pass
       
```

```python
lista_precos = []

# Para os números de 0 a 9
for empresas in range(10):
    try:
        seletor = f'#\\3{empresas} > td:nth-child(3) > span'
        dado_preco = navegador.find_element(By.CSS_SELECTOR, seletor).text
        lista_precos.append(dado_preco)
        print(dado_preco)
    except:
        pass

# Para os números de 10 a 99
for empresas in range(10, 100):
    try:
        seletor = f'#\\3{empresas // 10} {empresas % 10} > td:nth-child(3) > span'
        dado_preco = navegador.find_element(By.CSS_SELECTOR, seletor).text
        lista_precos.append(dado_preco)
        print(dado_preco)
    except:
        pass


```
